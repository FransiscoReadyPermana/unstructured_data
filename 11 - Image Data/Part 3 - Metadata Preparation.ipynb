# Mendapatkan Semua Image Path

```python
from glob import glob
```

```python
fname = glob('data/*/*.png')
fname[-5:]
```

# X = path => image => Flatten

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = []
for f in fname:
    x.append(plt.imread(f).flatten())
x = np.array(x)
```

```python
x
```

```python
X = np.array([plt.imread(f).flatten() for f in fname])
X
```

```python
X.shape
```

# y = Path => split => [1]

```python
y = np.array([int(f.split('\\')[1]) for f in fname])
y
```

```python
y.shape
```

# join jd satu dataframe

```python
df = pd.DataFrame(X, columns=range(1,785))
df
```

```python
df['label'] = y

df.to_csv('data/metadataExercise.csv', index=False)
```
